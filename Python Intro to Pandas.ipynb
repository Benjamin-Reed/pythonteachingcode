```python
xlr=pd.ExcelWriter(os.path.join(filePath,'babynames_by_year.xlsx'))  #create a placeholder for an Excel file
```

```python
for y in years:  #loop through each year
    criteria=f"year=={y}"  #create a criteria for each year
    print(criteria)
    df_temp=df.query(criteria) #create a temporary dataframe for each year
    df_temp.to_excel(xlr,str(y),index=False)  #push the temporary dataframe to a tab in the Excel file
xlr.close()   #save the Excel file
```
